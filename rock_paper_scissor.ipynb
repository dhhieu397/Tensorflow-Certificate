```python
import tensorflow as tf

def normalize(image, label):

  image = tf.image.resize(image, [128, 128])
  
  # Chun ha nh
  normalized_image = tf.cast(image, tf.float32) / 255.
  
  return normalized_image, label

# Thc hin trn b train

ds_train = ds_train.map(normalize)
ds_train = ds_train.batch(128)
ds_train_shuffled = ds_train.shuffle(100)

# Thc hin trn b validation

ds_valid = ds_valid.map(normalize)
ds_valid = ds_valid.batch(128)
ds_valid_shuffled = ds_valid.shuffle(100)
```
